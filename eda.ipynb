```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/intern/car.csv")
```

```python
df.head()
```

```python
# Fill missing values
df['mileage(km/ltr/kg)'].fillna(df['mileage(km/ltr/kg)'].median(), inplace=True)
df['engine'].fillna(df['engine'].median(), inplace=True)
df['seats'].fillna(df['seats'].median(), inplace=True)
df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')
df['max_power'].fillna(df['max_power'].median(), inplace=True)
```

```python
df.head()
```

```python
# 1. Distribution plots for numerical features
num_cols = ['year', 'selling_price', 'km_driven', 'mileage(km/ltr/kg)', 'engine', 'max_power', 'seats']
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# 2. Count plots for categorical features
cat_cols = ['fuel', 'seller_type', 'transmission', 'owner']
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="Set2")
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)")
plt.tight_layout()
plt.show()
```

```python
# 4. Boxplots of selling price by categorical columns
for col in cat_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=col, y='selling_price', data=df, palette="Set3")
    plt.title(f'Selling Price by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
